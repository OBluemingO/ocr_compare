dash_detail
1. กรองสีฟ้า
2. กรองสีเเดง
3. นำ 2 ภาพมารวมกัน
4. ทำให้ภาพให้อยู่สีตรงข้าม
ุ5. เกลี่ย noise ภาพให้ ข้อความชัดขึ้น

```python
def dash_detail(img):
  # filter color blue
  img_color_extract_blue = cv2.inRange(img, np.array([122,0,0]), np.array([255,255,255]))
  # filter color red
  img_color_extract_red = cv2.inRange(img, np.array([0,0,127]), np.array([255,255,255]))
  # merge both of them into one image
  merge_img = cv2.addWeighted(img_color_extract_blue, 0.6, img_color_extract_red, 0.6, 1)
  # make image to different color
  diff_img = cv2.bitwise_not(merge_img)
  # remove some noise we don't need it
  pre_img_result = cv2.adaptiveThreshold(diff_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 50)

  return pre_img_result
```

dash_statistics
1. กรองสีฟ้า
2. กรองสีเเดง
3. นำ 2 ภาพมารวมกัน
4. ทำให้ภาพให้อยู่สีตรงข้าม
5. กำหนด filter ที่เอาไว้ใช้กรอง
6. กรองรูปทำให้ font ใหญ่ขึ้น
7. เกลี่ย noise ภาพให้ ข้อความชัดขึ้น

```python
def dash_statistics(img):
  # filter color blue
  img_color_extract_blue = cv2.inRange(img, np.array([127,0,0]), np.array([255,255,255]))
  # filter color red
  img_color_extract_red = cv2.inRange(img, np.array([0,0,127]), np.array([255,255,255]))
  # merge both of them into one image
  merge_img = cv2.addWeighted(img_color_extract_blue, 0.75, img_color_extract_red, 0.75, 1)
  # make image to different color
  diff_img = cv2.bitwise_not(merge_img)
  # create kernal or filter image
  kernal_clear_noise = cv2.getStructuringElement(cv2.MORPH_RECT, (1,1))
  # make image bold font bigger 
  diff_img_font_bigger = cv2.erode(diff_img, kernal_clear_noise)
  # make image darker
  _, pre_img_result = cv2.threshold(diff_img_font_bigger,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

  return pre_img_result
```

```python
def cleaning_text(text, step, groups, teamA, teamB):
  if step == 0:
    # case name player
    result_cleaning = None
    # filter text if text in case [ japan, thai, eng ] 
    filter_text = re.findall("[a-zA-Z\u3040-\u30ff\u3400-\u4dbf\u4e00-\u9fff\uf900-\ufaff\uff66-\uff9fก-๙]+",text)
    # check if filter text not join each other then make text been one sentence
    if len(filter_text) > 1:
      result_cleaning = ''.join(filter_text)
    else:
      if len(filter_text) == 0:
        result_cleaning = ''
      else: 
        result_cleaning = filter_text[0]
    
    # if in teamA , else teamB then add cleaning data into dict
    if(groups <= 4):
      for index, key in enumerate(teamA):
         if index == step:
            teamA[key].append(result_cleaning)
    else:
      for index, key in enumerate(teamA):
          if index == step:
              teamB[key].append(result_cleaning)

  elif step > 0 and step < 5 or step in [6,8,10] :
    # filter only number or integer
    result_cleaning = re.findall("\d+",text)
    if len(result_cleaning) > 0:
      result_cleaning = result_cleaning[0]

    # if in teamA , else teamB then add cleaning data into dict
      if(groups <= 4):
        for index, key in enumerate(teamA):
          if index == step:
              teamA[key].append(result_cleaning)
      else:
        for index, key in enumerate(teamA):
            if index == step:
                teamB[key].append(result_cleaning)
    else:
        if(groups <= 4):
          for index, key in enumerate(teamA):
            if index == step:
                teamA[key].append('')
        else:
          for index, key in enumerate(teamA):
              if index == step:
                  teamB[key].append('')
  else:
    cleaning_text_result = re.findall("[\d+\%]",text)
    result_cleaning = ''
    # if text don't have % and text not emtry then add % to last position
    if '%' not in cleaning_text_result and len(cleaning_text_result) > 0:
      cleaning_text_result.append('%')
    else:
      for index, value in enumerate(cleaning_text_result):
        # if % not in last position then move it to last position
        if value == '%' and index+1 != len(cleaning_text_result):
          cleaning_text_result.pop(index)
          cleaning_text_result.append('%')
    # if text more than 4 charator that can't be cuz data we need in this case is 4 charactor if more then 4 it must been [] but ocr engine is terrible then we need to remove it from data
    if len(cleaning_text_result) > 4:
      cleaning_text_result.pop(0)

    # ocr can't read . in my logic then we need to add . into data ,just get position of we need and then add it into position we need and convert to float  
    if(len(cleaning_text_result) > 0):
      cleaning_text_result.insert(-2, ".")
      pure_text = ''.join(cleaning_text_result[:-1:])
      result_cleaning = float(pure_text)

    # if in teamA , else teamB then add cleaning data into dict
    if(groups <= 4):
      for index, key in enumerate(teamA):
         if index == step:
            teamA[key].append(result_cleaning)
    else:
      for index, key in enumerate(teamA):
         if index == step:
            teamB[key].append(result_cleaning)
    
```

# ปรับขนาดของ array ทั้ง 2 ตัวให้มีขนาดเท่ากัน

```python
# make list is balance
def make_list_balance( list_origin, list_ocr_new, not_count_origin, not_count_ocr_new ):
  # notCount = None
  if len(list_origin) > len(list_ocr_new):
    # notCount = not_count_ocr_new
    range = len(list_origin) - len(list_ocr_new)
    count = 0
    while range != count :
      list_ocr_new.append('')
      count += 1
  else:
    # notCount = not_count_origin
    range = len(list_ocr_new) - len(list_origin) 
    count = 0
    while range != count :
      list_origin.append('')
      count += 1
```

# ทำการเเยกประเภทว่า ข้อมูล เป็นข้อมูลประเภทไหนแล้วเก็บใส่ตัวแปร
 ประเภทข้อมูลก็จะได้เเก่
 [คำที่ถูก, คำที่เปลี่ยน, คำที่ถูกลบ, คำที่ถูกเพิ่ม]

```python
def find_match_origin(range_origin, list_origin, list_ocr_new, total_origin, total_ocr, not_count, ocr_engine, check_column):
  notMatch = []
  match = []
  removeWord = total_origin - total_ocr
  insertWord = total_ocr - total_origin

  increase = 0
  looping = True

  while looping:
    if range_origin == increase :
      looping = False
    if list_origin[increase] != list_ocr_new[increase]:
      if increase <=  not_count: 
        notMatch.append(increase)
    else:
      match.append(increase)
    increase += 1

  if insertWord < 0:
    insertWord = 0
  if removeWord < 0:
    removeWord = 0

  top = len(notMatch) + insertWord + removeWord
  bottom = len(notMatch) + insertWord + removeWord + len(match)

  cer_normalization  = top / bottom
  if ocr_engine == 'py':
    average_percentage_py.append(100-(cer_normalization*100))
    if check_column == 0:
      name_average_percentage_py.append(100-(cer_normalization*100))
    else:
      not_name_average_percentage_py.append(100-(cer_normalization*100))
  else:
    average_percentage_easy.append(100-(cer_normalization*100))
    if check_column == 0:
      name_average_percentage_easy.append(100-(cer_normalization*100))
    else:
      not_name_average_percentage_easy.append(100-(cer_normalization*100))
```

#main function ที่เอาไว้เรียกเพื่อหาสูตร CER

```python
def find_cer_word(original_data, data_py=None, data_easy=None):
   for index_root, key in enumerate(original_data):
    for index_root_second, column in enumerate(original_data[key]):
      for index_root_third, key_inside in enumerate(original_data[key][column]):
        origin_value = None
        py_value = None
        easy_value = None
        for index, value in enumerate(original_data[key][column][key_inside]):
          if isinstance(original_data[key][column][key_inside][index], str):
            origin_value =list(original_data[key][column][key_inside][index].lower())
            py_value =list(data_py[key][column][key_inside][index].lower())
            easy_value = list(data_easy[key][column][key_inside][index].lower())

            origin_not_count = len(origin_value) - 1
            py_not_count = len(py_value) - 1
            easy_not_count = len(easy_value) - 1
            origin_total = len(origin_value) 
            py_total = len(py_value)
            easy_total = len(easy_value)

            make_list_balance(origin_value, py_value, origin_not_count, py_not_count)
            make_list_balance(origin_value, easy_value, origin_not_count, easy_not_count)

            find_match_origin(origin_not_count, origin_value, py_value, origin_total, py_total, py_not_count, 'py', index_root_third)
            find_match_origin(origin_not_count, origin_value, easy_value, origin_total, easy_total, easy_not_count, 'easy', index_root_third)

          else:
            origin_value = list(str(original_data[key][column][key_inside][index]))
            py_value = list(str(data_py[key][column][key_inside][index]))
            easy_value = list(str(data_easy[key][column][key_inside][index]))

            origin_not_count = len(origin_value) - 1
            py_not_count = len(py_value) - 1
            easy_not_count = len(easy_value) - 1
            origin_total = len(origin_value) 
            py_total = len(py_value)
            easy_total = len(easy_value)

            make_list_balance(origin_value, py_value, origin_not_count, py_not_count)
            make_list_balance(origin_value, easy_value, origin_not_count, easy_not_count)

            find_match_origin(origin_not_count, origin_value, py_value, origin_total, py_total, py_not_count, 'py', index_root_third)
            find_match_origin(origin_not_count, origin_value, easy_value, origin_total, easy_total, easy_not_count, 'easy', index_root_third)
```

```python
def create_excel(data, name):
  for sheet in data:
    for team in data[sheet]:
      name_sheet = sheet + '_' +team
      covert_to_excel = data[sheet][team]
      try:
        with pd.ExcelWriter(name, engine='openpyxl', mode="a") as writer:  
          df = pd.DataFrame.from_dict(covert_to_excel)
          df.to_excel(writer, sheet_name=name_sheet)
      except:
        df = pd.DataFrame.from_dict(covert_to_excel)
        df.to_excel(name, sheet_name=name_sheet)
```
